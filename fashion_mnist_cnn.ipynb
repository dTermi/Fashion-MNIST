```python
import torchvision
import torchvision.transforms as transforms
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
train_set = torchvision.datasets.FashionMNIST(root = ".", train=True, download=True, transform=transforms.ToTensor())
test_set = torchvision.datasets.FashionMNIST(root = ".", train=False, download=True, transform=transforms.ToTensor())
train_loader = torch.utils.data.DataLoader(train_set, batch_size=32, shuffle=True) # Shuffle to keep the NN from remembering the data and its position
test_loader = torch.utils.data.DataLoader(test_set, batch_size=32, shuffle=False)
# Fix the seed to get the same randomness for each run
torch.manual_seed(0)
```

```python
# If using CuDNN
torch.cuda.manual_seed(0)
torch.backends.cudnn.deterministic=True
torch.backends.cudnn.benchmark=False
```

```python
'''
Get one image from the training data loader and its label.
reshape the image so that it is a square and in greyscale and plot the image
'''
input_data, label = next(iter(train_loader))
plt.imshow(input_data[0, :, :, :].numpy().reshape(28, 28), cmap='gray_r', label = label[0])
plt.tight_layout()
plt.show()
```

```python
class MNIST_Conv(nn.Module):
    '''
    I'm taking the activation function and the drop out as a parameter 
    when building the convolution layers.

    The model will be trained as per the activation function passed as a parameter
    and the drop out will be applied if boolean True is Passed.
    '''
    def __init__(self, activation_function, drop_out=False):
        super(MNIST_Conv, self).__init__()

        # Calc Output dimension using the formula from the slides: (26-2+2)/2 x (26-2+2)/2 x 12 = 13 x 13 x 12
        # Last number of the op dimension is the output of that layer and input of the next layer

        # Convolutional Layers
        '''
        1 input image and 32 output channels
        5x5 kernel size
        '''
        self.conv1 = nn.Conv2d(1, 32, kernel_size=5, stride=1)
        self.activ1 = activation_function # The selected activation function
        self.max_pool1 = nn.MaxPool2d(2, stride=2) # 

        '''
        32 input channels and 64 output channels
        5x5 kernel size
        '''
        self.conv2 = nn.Conv2d(32, 64, kernel_size=5, stride=1)
        self.activ2 = activation_function
        self.max_pool2 = nn.MaxPool2d(2, stride=2)

        # Fully Connected Layers 1, 2 and 3.
        self.full_conn_1 = nn.Linear(1024, 1024)
        self.act1 = activation_function
        self.full_conn_2 = nn.Linear(1024, 256)
        self.act2 = activation_function
        # IF the drop out boolean is Tue, add the drop out functionality.
        if drop_out == True:
            self.drop1 = nn.Dropout(0.3)
            self.full_conn_3 = nn.Linear(256, 10)
        else:
            self.full_conn_3 = nn.Linear(256, 10)

    def forward(self, x, drop_out):
        # Pass the input through the CNN Layers.
        x = self.conv1(x)
        x = self.activ1(x)
        x = self.max_pool1(x)

        x = self.conv2(x)
        x = self.activ2(x)
        x = self.max_pool2(x)

        x = x.view(x.size(0), -1) # Flatten the output for fully connected layers

        # Pass the input through the Fully Connected Layers.
        x = self.full_conn_1(x)
        x = self.act1(x)
        x = self.full_conn_2(x)
        x = self.act2(x)
        if drop_out == True:
            x = self.drop1(x)
        x = self.full_conn_3(x)
        
        return x
```

```python
def eval(dataloader, drop_out):
    total, correct = 0, 0
    # Enable the evaluation mode for our model.
    CNN.eval()
    # Get data from the dataloader.
    for data in dataloader:
        inputs, labels = data # Separate the image and its label
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = CNN(inputs, drop_out) # Pass the training set to the forward function along with the dropout boolean
        _, pred = torch.max(outputs.data, 1) # pred takes the predicted class value where the maximum values occurred
        total += labels.size(0)
        correct += (pred == labels).sum().item()
    return 100 * correct / total # Returns the accuracy
```

```python
def init_weights(m):
    if isinstance(m, nn.Linear) or isinstance(m, nn.Conv2d):
        torch.nn.init.xavier_uniform_(m.weight)
```

```python
# Check if GPU is available and set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
alpha = 0.1 # Learning Rate
activation = nn.ReLU() # Activation function
Drp_out = False # If drop out needs to be implemented

CNN = MNIST_Conv(activation, Drp_out).to(device) # Init the CNN model and pass the activation function and drop out flag
# CNN = MNIST_Conv().to(device)
CNN.apply(init_weights) # apply the weights to the model

loss_func = nn.CrossEntropyLoss() # Set the loss function
loss_func.to(device)

optimization = torch.optim.SGD(list(CNN.parameters()), lr = alpha) # Select the optimization function

loss_epoch_array = [] # loss per epoch
max_epochs = 30 # Max epochs
loss_epoch = 0 # Loss for each epoch
train_accuracy = [] # List for all the training accuracy
test_accuracy = [] # List for all the testing accuracy

# Loop through each epoch
for epoch in range(max_epochs):
    loss_epoch = 0 # Set value of loss epoch to zero for each iteration
    for i, data in enumerate(train_loader, 0): 
        CNN.train() # Train the model
        inputs, labels = data # Separate the components of data
        inputs, labels = inputs.to(device), labels.to(device)
        optimization.zero_grad() # Clear gradients from previous iteration
        outputs = CNN(inputs, drop_out=Drp_out) # Forward Pass
        loss = loss_func(outputs, labels) # Calculate the loss
        loss.backward() # Backpropagation
        optimization.step() # Update the weights
        loss_epoch += loss.item() # Add all the losses for each epochs

    loss_epoch_array.append(loss_epoch) # Store all the losses
    train_accuracy.append(eval(train_loader, drop_out=Drp_out)) # Calculate the Training accuracy
    test_accuracy.append(eval(test_loader, drop_out=Drp_out)) # Calculate the Testing accuracy
    print("Epoch {}: Loss: {}, Train accuracy: {}, Test accuracy:{}".format(epoch + 1, loss_epoch_array[-1], train_accuracy[-1], test_accuracy[-1])) # Display the loss and accuracy for each epoch
```

```python
# Plot the plots
fig = plt.figure(figsize=(12, 6)) 
fig.add_subplot(1, 2, 1)
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.plot(train_accuracy, "r")
plt.plot(test_accuracy, "b")
plt.gca().legend(('train','test'))
plt.title('Train-Test Accuracy:: Alpha = 0.1, Activation = Relu')

# fig = plt.figure(figsize=(10, 7)) 
fig.add_subplot(1, 2, 2) 
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.plot(loss_epoch_array)
plt.gca().legend(('loss'))
plt.title('Loss:: Alpha = 0.1, Activation = Relu')

plt.show()
```
